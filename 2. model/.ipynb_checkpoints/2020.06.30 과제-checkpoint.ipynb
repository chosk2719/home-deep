```python
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.datasets import load_breast_cancer
from sklearn import datasets

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
seed = 2020
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
data = datasets.load_breast_cancer()
```

```python
x = data.data
y = data.target
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=seed)
```

```python
# 모델 설정
model = Sequential([
    Dense(300, input_dim=30, activation='relu'),
    Dense(12, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
]) 
model.summary()
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=2000,
          batch_size=100)
```

```python
# early_stopping_callback = EarlyStopping(monitor='loss', patience=100)
```

```python
# history = model.fit(x_train, y_train, validation_split=0.2, epochs=2000, batch_size=200, 
#           verbose=1, callbacks=[early_stopping_callback])
```

```python
y_loss=history.history['loss']
```

```python
y_acc=history.history['accuracy']
```

```python
x_len = np.arange(len(y_acc))
plt.figure(figsize=(12,8))
plt.ylim((0,1))
plt.plot(x_len, y_loss, "o", c="red", markersize=2)
plt.plot(x_len, y_acc, "o", c="blue", markersize=2)

plt.show()
```

```python
print("\n Accuracy: %.4f" % (model.evaluate(x_train, y_train, verbose=2)[1]))
```

```python
early_stopping_callback = EarlyStopping(monitor='loss', patience=100)
```

```python
history = model.fit(x_train, y_train, validation_split=0.2, epochs=2000, batch_size=200, 
          verbose=1, callbacks=[early_stopping_callback])
```

```python
y_loss=history.history['loss']
y_acc=history.history['accuracy']
y_vloss=history.history['val_loss']
y_vacc=history.history['val_accuracy']

x_len = np.arange(len(y_acc))
plt.figure(figsize=(12,8))
plt.plot(x_len, y_loss, "o", c="red", markersize=2)
plt.plot(x_len, y_acc, "o", c="orange", markersize=2)
plt.plot(x_len, y_vloss, "o", c="green", markersize=2)
plt.plot(x_len, y_vacc, "o", c="blue", markersize=2)

plt.show()
```

```python
print("\n Accuracy: %.4f" % (model.evaluate(x_train, y_train, verbose=2)[1]))
```

```python
from sklearn.datasets import load_digits
```

```python
di = load_digits()
di['data']
```

```python
x = di.data
y = di.target
```

```python
from sklearn.preprocessing import LabelEncoder

l = LabelEncoder()
l.fit(y)
y = l.transform(y)
```

```python
y
```

```python
Y_encoded = tf.keras.utils.to_categorical(y)
Y_encoded
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=seed)
```

```python
model = Sequential([
    Dense(300, input_dim=64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(28, activation='relu'),
    Dense(12, activation='relu'),
    Dense(1, activation='softmax')
]) 
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
import os
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
```

```python
modelpath = MODEL_DIR + "final{epoch:03d}-{val_loss:.4f}.hdf5"

checkpointer_callback = ModelCheckpoint(filepath=modelpath, monitor='val_loss', 
                               verbose=1, save_best_only=True)
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=100)
```

```python
history = model.fit(x_train, y_train, validation_split=0.2, epochs=2000, batch_size=100, 
          verbose=1, callbacks=[early_stopping_callback])
```

```python
history = model.fit(x_train, y_train, validation_split=0.2, epochs=2000, batch_size=500,
                    verbose=1, callbacks=[early_stopping_callback, checkpointer_callback])
```

```python
del model
```

```python
print("\n Accuracy: %.4f" % (model.evaluate(x_test, y_test, verbose=2)[1]))
```
